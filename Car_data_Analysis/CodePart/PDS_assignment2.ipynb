```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv("/content/train.csv")  # Replace with actual path
df
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].median())
df['Engine'] = df['Engine'].fillna(df['Engine'].median())
df['Power'] = df['Power'].fillna(df['Power'].median())
df['New_Price'] = df['New_Price'].fillna(df['New_Price'].mode()[0])
df['Fuel_Type'] = df['Fuel_Type'].fillna(df['Fuel_Type'].mode()[0])
df['Transmission'] = df['Transmission'].fillna(df['Transmission'].mode()[0])
```

```python

# Remove unit strings BEFORE trying to fillna or convert
df['Mileage'] = df['Mileage'].astype(str).str.replace(' kmpl', '', regex=False)
df['Mileage'] = df['Mileage'].astype(str).str.replace(' km/kg', '', regex=False)
df['Mileage'] = pd.to_numeric(df['Mileage'], errors='coerce')

df['Engine'] = df['Engine'].astype(str).str.replace(' CC', '', regex=False)
df['Engine'] = pd.to_numeric(df['Engine'], errors='coerce')

df['Power'] = df['Power'].astype(str).str.replace(' bhp', '', regex=False)
df['Power'] = pd.to_numeric(df['Power'], errors='coerce')

df['New_Price'] = df['New_Price'].astype(str).str.replace(' Lakh', '', regex=False)
df['New_Price'] = pd.to_numeric(df['New_Price'], errors='coerce')
```

```python
# Now you can safely fill missing numerical values
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].median())
df['Engine'] = df['Engine'].fillna(df['Engine'].median())
df['Power'] = df['Power'].fillna(df['Power'].median())
df['New_Price'] = df['New_Price'].fillna(df['New_Price'].mode()[0])

# And for categorical columns
df['Fuel_Type'] = df['Fuel_Type'].fillna(df['Fuel_Type'].mode()[0])
df['Transmission'] = df['Transmission'].fillna(df['Transmission'].mode()[0])
```

```python
# Remove units like 'kmpl', 'CC', 'bhp', 'Lakh'
df['Mileage'] = df['Mileage'].astype(str).str.replace(' kmpl', '').astype(float)
df['Engine'] = df['Engine'].astype(str).str.replace(' CC', '').astype(float)
df['Power'] = df['Power'].astype(str).str.replace(' bhp', '').astype(float)
df['New_Price'] = df['New_Price'].astype(str).str.replace(' Lakh', '').astype(float)
```

<!-- JUSTIFICATION -->

<!-- JUSTIFICATION  -->
I didn’t want to lose a lot of data, so instead of removing rows, I filled in the missing values.

For the number-based columns, I used the median because it’s more stable—even if there are some very high or very low values, it won’t get affected much.

And for columns with text (like fuel type or transmission), I just used the most common value (mode), because that feels like the safest guess.

I only thought about dropping a column if it was missing a lot of data and wasn’t super important. This way, I could keep the dataset as complete and useful as possible.

```python
# Convert Fuel_Type and Transmission to numeric format
df = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission'], drop_first=True)
```

```python
from datetime import datetime
current_year = datetime.now().year

# Create a new column "Car_Age"
df['Car_Age'] = current_year - df['Year']
```

```python
# Split 'Name' into 'Make' and 'Model'
df['Make'] = df['Name'].str.split().str[0]
df['Model'] = df['Name'].str.split().str[1:].str.join(' ')

# Rename 'Price' to 'Car_Price'
df.rename(columns={'Price': 'Car_Price'}, inplace=True)

# Select: only Make, Model, and Car_Price
selected = df[['Make', 'Model', 'Car_Price']]
print("Selected Columns:\n", selected.head())

# Filter: show only Diesel cars if 'Fuel_Type_Diesel' exists
if 'Fuel_Type_Diesel' in df.columns:
    diesel_cars = df[df['Fuel_Type_Diesel'] == 1]
    print("Diesel Cars:\n", diesel_cars[['Make', 'Model', 'Car_Price']].head())
else:
    print("Column 'Fuel_Type_Diesel' not found.")

# Mutate: create a new column dividing price by engine size
df['Price_per_CC'] = df['Car_Price'] / df['Engine']

# Arrange: sort by Car_Price from high to low
sorted_df = df.sort_values(by='Car_Price', ascending=False)
print("Sorted by Car_Price:\n", sorted_df[['Make', 'Model', 'Car_Price']].head())

# Group by Fuel_Type_Diesel and get average price if the column exists
# Create a new column to indicate Diesel cars if not already present
if 'Fuel_Type_Diesel' not in df.columns:
    df['Fuel_Type_Diesel'] = (df[['Fuel_Type_Electric', 'Fuel_Type_Petrol']].sum(axis=1) == 0).astype(int)

```

```python
df.to_csv('cleaned_used_car_data.csv', index=False)
```

```python
# Group by Diesel status and calculate average car price
grouped_avg_price = df.groupby('Fuel_Type_Diesel')['Car_Price'].mean().reset_index()

# Replace 0/1 with readable labels
grouped_avg_price['Fuel_Type'] = grouped_avg_price['Fuel_Type_Diesel'].map({0: 'Not Diesel', 1: 'Diesel'})

# Reorder columns
grouped_avg_price = grouped_avg_price[['Fuel_Type', 'Car_Price']]

# Rename for clarity
grouped_avg_price.columns = ['Fuel_Type', 'Average_Car_Price']

# Print the nicely formatted result
print("\nAverage Car Price by Fuel Type:\n")
print(grouped_avg_price)
```

```python
# Create readable labels from dummy columns
df['Fuel_Type_Label'] = df[['Fuel_Type_Diesel', 'Fuel_Type_Petrol', 'Fuel_Type_Electric']].idxmax(axis=1).str.replace('Fuel_Type_', '')
df['Transmission_Label'] = df['Transmission_Manual'].apply(lambda x: 'Manual' if x == 1 else 'Automatic')

# Group by the readable labels and calculate average price
grouped_summary = df.groupby(['Fuel_Type_Label', 'Transmission_Label'])['Car_Price'].mean().reset_index()

# Print the summary
print("\nAverage Car Price based on Fuel Type and Transmission:\n")
print(grouped_summary)

# Save to CSV
grouped_summary.to_csv("task_e_groupby_fuel_trans_summary.csv", index=False)
```

```python
# Create readable labels from one-hot encoded columns
df['Fuel_Type_Label'] = df[['Fuel_Type_Diesel', 'Fuel_Type_Petrol', 'Fuel_Type_Electric']].idxmax(axis=1).str.replace('Fuel_Type_', '')
df['Transmission_Label'] = df['Transmission_Manual'].apply(lambda x: 'Manual' if x == 1 else 'Automatic')

# Group by fuel and transmission labels
grouped_avg = df.groupby(['Fuel_Type_Label', 'Transmission_Label'])['Car_Price'].mean().reset_index()

# Show the result
print("\nFuel Type and Transmission-wise Average Price:\n")
print(grouped_avg)

# Save to CSV
grouped_avg.to_csv("task_e_groupby_fuel_trans_summary.csv", index=False)
```
